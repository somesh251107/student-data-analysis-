```python
# *Fix Invalid Data:**

#Remove or fix rows with negative total_amount
df[df["total_amount"]<0]
```

```python
# Ensure discount_percent is between 0 and 100
df[(df["discount_percent"]<0) | (df["discount_percent"]>100)]
```

```python
df["unit_price"] = df["total_amount"] / df["quantity"]
```

```python
df["unit_price"]
```

```python
#   - `net_amount`: total_amount after applying discount
# inf
df["net_amount"] = df["total_amount"] * (1 - df["discount_percent"].fillna(0) /100)
```

```python
df["net_amount"].isnull().sum()
```

```python
df["net_amount"].fillna(0)
```

```python
#   - `year_month`: Extract year-month from purchase_date
df["year_month"] = df["purchase_date"].dt.to_period("M")
```

```python
df["year_month"].head()
```

```python
df.head()
```

```python
'''customer_segment`: Categorize customers as:
     - "High Value" if total spending > $500
     - "Medium Value" if total spending $100-$500
     - "Low Value" if total spending < $100'''
```

```python
customer_total = df.groupby("customer_id")["net_amount"].sum()
```

```python
df["total_spending"] = df["customer_id"].map(customer_total)
```

```python
df.head()
```

```python
df["customer_segment"] = np.select(
    [
     df["total_spending"] > 500,
     (df["total_spending"] >= 100) & (df["total_spending"]<=500),
     df["total_spending"]<100
    ],
    ["High Value","Medium Value", "Low Value"],
    default="unknown"
)
```

### **Phase 5: Data Visualization (Bonus)**

```python
#Create visualizations using matplotlib or seaborn:
 #   - Bar chart: Sales by Product Category
category_sales = df.groupby("product_category")["net_amount"].sum()
category_sales.plot(kind="bar")

plt.title("sales by product category")
plt.xlabel("product categoory")
plt.ylabel("total sale")


plt.show()
```

```python
# Pie chart: Market share by payment method
payments = df.groupby("payment_method")["net_amount"].sum()
payments.plot(kind="pie", y="net_amount",autopct="%1.1f%%")

plt.title("payment method")
```

```python
# Histogram: Distribution of customer ages
plt.hist(df["customer_age"].dropna(),bins=10)

plt.title("customer age distribution")
plt.xlabel("customer age")
plt.ylabel("frequency")

plt.show()
```

```python
#Line plot: Sales trend over time
sales = df.groupby("purchase_date")["net_amount"].sum()
sales.plot(kind="line")

plt.title("sales trands over time")
plt.xlabel("date of purchaing")
plt.ylabel("total amount")
```

```python
# Box plot: Order value by product category
sales= df["product_category"].unique()
products = [df[df["product_category"]==cat]["net_amount"].dropna() for cat in sales]

plt.boxplot(products , label=sales)

plt.title("order value by product category")
plt.xlabel("product category")
plt.ylabel("order value")


plt.show()
```
